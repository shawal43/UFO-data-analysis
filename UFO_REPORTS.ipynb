```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.tail()
```

```python
ufo = ufo.dropna(how='any')
```

```python
ufo = ufo.drop_duplicates()
```

```python
ufo.head()
```

```python
ufo['Time'] = pd.to_datetime(ufo.Time)
```

```python
ufo['year'] = ufo.Time.dt.year
```

```python
ufo['time_only'] = ufo.Time.dt.time
```

```python
ufo.head()
```

```python
ufo = ufo.drop(columns=['Time'])
```

```python
ufo.head()
```

```python
ufo['Shape Reported'].value_counts()
```

```python
Common_shape = ufo['Shape Reported'].value_counts()
```

```python
Common_shape.plot(kind='bar' , figsize=(20, 10) , title=('COMMON SHAPE OF UFO'))
```

```python
peak_year = ufo.year.value_counts().sort_index()
```

```python
peak_year.plot(kind='bar' , figsize=(20 ,10) , title=('ufo sightings by year'))
```

```python
common_cities = ufo['City'].value_counts().head(10)
```

```python
ufo.City.value_counts()
```

```python
common_cities.plot(kind='bar', figsize=(20 ,10) , title=('ufo sightings by cities'))
```

```python
common_ufo_color = ufo['Colors Reported'].value_counts().head()
```

```python
common_ufo_color.plot(kind='bar' , figsize=(20 ,10) , title=('ufo sightings by colour'))
```
